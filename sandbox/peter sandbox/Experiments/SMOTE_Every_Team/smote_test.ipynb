```python
results = []
print(f"team,accuracy,AUC,recall,prec,f1,kappa,MCC")
for team in teams:
    try:
        
        try:
            dataset = pd.read_parquet(f"data/pp_all_teams/{team['LKUPCLIENTID']}.parquet")
            print(f"read {team['TEAM']} from {team['LKUPCLIENTID']}.parquet")
        
        except:
            print("Failed to read parquet, reading SQL")
            time.sleep(60)
            dataset = helpers.get_product_propensity_dataset(
                cluster=team["CLUSTER"],
                database=team["DATABASE"],
                lkupclientid=team["LKUPCLIENTID"],
                scoring_year=team["SCORING_YEAR"],
                type_flag=0 
            )
            print("Read from sql")
            dataset.to_parquet(f"data/pp_all_teams/{team['LKUPCLIENTID']}.parquet")


        dataset.shape
        df = dataset.copy()
        # choose the features for the stellar base PP model
        features = [
            "atp_last",
            "attendedpct_last",
            "attendedpct_prior",
            "attendance_current",
            "attended_last",
            "attended_prior",
            "clientcode",
            "dimcustomermasterid",
            "date_last_save",
            "date_last_send",
            # "date_last_touch",
            "events_last",
            "events_prior",
            "distance",
            # "distance_buckets",
            "events_current",
            "lkupclientid",
            "opens",
            "opens_prior",
            "product",
            "product_current",
            "product_last",
            "seasonyear",
            "sends",
            "sends_prior",
            "spend_current",
            # "spend_current_buckets",
            "tenure",
            "volume_current"
        ]

        # copy your main dataframe
        df_dataset = df

        # creating the training and evaluation dataframe
        df_dataset = df_dataset[features]

        # limiting classes to only product plans
        product_plans = team["PRODUCT_PLANS_ALLOWED"]

        df_dataset = df_dataset[df_dataset["product"].isin(product_plans)]
        print(f"Before smote size: {df_dataset.shape}")
        # df_dataset = df_dataset.fillna(method="ffill")
        # df_dataset = df_dataset.dropna()

        # cat_idx_lst = [index for index, col in enumerate(df_dataset.columns) if df_dataset[col].dtype.name == 'object']
        # sm = SMOTENC(random_state=42, categorical_features=cat_idx_lst)

        # X_res, y_res = sm.fit_resample(df_dataset.drop(["product"], axis=1), df_dataset["product"])

        # df_dataset = X_res
        # df_dataset["product"] = y_res
        print(f"After smote size: {df_dataset.shape}")

        df_train = df_dataset.sample(frac=0.90, random_state=786)
        df_eval = df_dataset.drop(df_train.index)

        df_train.reset_index(drop=True, inplace=True)
        df_eval.reset_index(drop=True, inplace=True)

        # print out the number of records for training and eval
        print('Data for Modeling: ' + str(df_train.shape))
        print('Unseen Data For Predictions: ' + str(df_eval.shape), end="\n\n")

        setup(
            data= df_train, 
            target="product", 
            train_size = 0.85,
            data_split_shuffle=True,
            silent=True,
            verbose=False,
            ignore_features=[
                # "attendedpct_prior",
                # "attendedpct_last",
                "attendance_current",
                "clientcode",
                "dimcustomermasterid",
                # "distance",
                "events_current",
                "lkupclientid",
                # "product_current",
                # "product_last",
                "seasonyear",
                "spend_current",
                "volume_current",
                "sends",
                "sends_prior",
                "opens",
                "opens_prior",
                "date_last_send",
                # "date_last_touch",
                "date_last_save"
            ],
            categorical_features=[
                "product_current",
                "product_last",
            ],
            numeric_features=[
                "atp_last",
                #"attendance_current",
                "attended_last",
                "attended_prior",
                "distance",
                "events_last",
                "events_prior",
                #"spend_current",
                "tenure"
            ],
        );

        model_matrix = compare_models(
            fold=10,
            include=["rf"]
        )

        best_model = create_model(model_matrix, cross_validation=True);

        accuracies = pull()
        accuracy = accuracies[accuracies.index=="Mean"]["Accuracy"].values[0]
        AUC = accuracies[accuracies.index=="Mean"]["AUC"].values[0]
        recall = accuracies[accuracies.index=="Mean"]["Recall"].values[0]
        prec = accuracies[accuracies.index=="Mean"]["Prec."].values[0]
        f1 = accuracies[accuracies.index=="Mean"]["F1"].values[0]
        kappa = accuracies[accuracies.index=="Mean"]["Kappa"].values[0]
        MCC = accuracies[accuracies.index=="Mean"]["MCC"].values[0]

        print(f"{team['TEAM']},{accuracy},{AUC},{recall},{prec},{f1},{kappa},{MCC}")
        results.append([team['TEAM'],accuracy,AUC,recall,prec,f1,kappa,MCC])
    except Exception as e:
        print(f"Skipping team {team['TEAM']} for exception: {e}")

result_df = pd.DataFrame(results, columns=["team","accuracy","AUC","recall","prec","f1","kappa","MCC"])
result_df.to_csv("before_smote_more-features.csv", index=False)
```
