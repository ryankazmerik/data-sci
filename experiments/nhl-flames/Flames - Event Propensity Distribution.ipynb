```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df_scores_1 = pd.read_csv("data/scores_2022-10-12.csv")
df_scores_2 = pd.read_csv("data/scores_2022-10-13.csv")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Event Propensity Scores')

sns.histplot(ax=axes[0], data=df_scores_1, x='sascore', bins= 20, kde=True)
axes[0].set_title("Old Scores")

sns.histplot(ax=axes[1], data=df_scores_2, x='sascore', bins= 20, kde=True)
axes[1].set_title("New Scores")
```
